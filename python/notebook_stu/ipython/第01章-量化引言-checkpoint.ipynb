## 1.1 什么是量化交易
<i>量化交易是指以先进的属性模型替代人为的主管判断，利用计算机技术从庞大的历史数据中海选出能带来超额收益的多种“大概率”事件以制定策略。它极大地降低了市场波动给投资者情绪带来的影响，避免在市场极度狂热或悲观的情况下做出非理性的投资决策。</i>

我们在生活中经常使用搜索功能，在搜索框中输入我们的问题，比如“极客营在哪里”,通过搜索我们就可以得到我们想知道的答案。

那么，如果我们通过搜索来知道“今天买什么股票能挣钱”呢？

搜索无法告诉你答案，而量化系统的任务就是类似的通知交易者今天该买什么，今天该卖什么。

普通的交易是由交易者根据自身经验或者偏好进行投资决策，量化交易是通过将数据（行情历史，基本面信息和新闻资讯等）输入量化模型后，利用计算机和统计学技术方法分析数据，产生交易信号进行交易决策。量化交易简单示意图如下：

量化模型一般包括选股模块，择时模块，风险控制模块。

传统的交易使用定性分析进行投资决策：
* 传统的交易决策不论是技术面分析还是基本面分析都属性定性分析；
* 定性分析一般是通过人的思维来完成，它的有点是在深度上占用绝对优势。

量化交易使用定量分析进行投资决策：
* 量化交易投资决策属于定量分析，它以历史数据分析为基础，利用数学、统计学等工具高效快速地进行决策；
* 定量分析利用计算机强大的运算能力，在广度上占用绝对优势。

定量分析也会使用到定性分析中的很多技术，大多情况是人的思维对定性分析的高度抽象形成模式，将模式应用到定量分析以提升广度。最理想的情况是在不丢失深度的情况下，完全将思想移植广度提升，这样在大数定理的魔力下，更强有力的战胜市场。

## 1.2 量化交易：投资？投机？赌博？
普遍来说，我们都希望自己的交易是投资，自己是投资者。很少有人接受自己的交易是投机，更不会有人希望自己的交易被称为赌博。

但现实很残酷，事实上大部分人的交易方式确实是赌博。他们不断的在市场进行短线交易，只在乎胜率。例如，他们经常会赌那个股票价格涨太多了，应该下跌了，或者那个股票价格下跌的太多了，应该上涨了。一旦走势没有如预期那样发展，他们则会将亏损的部分坚持持有，认为应长期持有，长线投资。

很多人分不清赌博和投机的区别，赌博靠的就是运气，完全无法获得概率优势。投机需要技巧和经验，权衡得失，获取概率优势。

量化交易中大多数策略是基于对历史规律的总结，在规律的基础上发现概率优势，它的最大理论依据是人性的相似性以及人性很难改变的事实。如果每一瞬间的股票价格都是全体交易者对价值所达成的一种瞬间的共识，那么历史的规律在今后的交易中同样具有指导意义。
由于世界上没有两片树叶是完全相同的，所以量化交易的概率优势并不具有绝对的优势，即达不到预测的程度。传统意义上的投资需要一个相对比较大的概率优势目标，也就是说投资的目标需要一个比较准确的预测结果。投资是指前期投入大量资源对投资目标进行研究分析，以获取丰厚投资回报为目标，进行长期投资的行为才能被称为投资。其他所有的简单分析，中短期持有的行为都应该属于投机范畴。作为个人小资金投资者，很难真正的做到基本面分析，所以我们的量化交易应该更倾向于投机的范畴。

很多交易者错误的认为自己是具有短线预测天赋的专家，但其实他们只是市场中的“赌徒”。比起赌徒来书，相对于大多数个人量化交易者或中小资金来说，通过量化交易技术获取交易概率优势，成为市场中成功的投机者，一步一步脱变为投机家也是个很好的选择。

## 1.3 量化交易的优势
* 量化交易通过计算机强大的运算能力，在市场广度分析上占有绝对优势；
* 量化交易通过历史规律的总结，在其基础上发现概率优势，形成良好投机基础。

下面为量化交易的其他优势
### 1.3.1 避免短线频繁交易
通过量化交易，上瘾型交易者可以将自身抽离决策前沿，只通过计算机发出的信号进行买卖。劳动型交易者也可以通过开发学习量化相关知识来避免短线频繁交易，但是要注意不要过度拟合交易系统，以及实现太过复杂的交易策略。
### 1.3.2 避免逆势操作
顺势交易可以说是交易者必须遵循的交易模式，但是大多数交易者喜欢逆势操作，比如只在股票下跌很多的时候进行买入。造成这种现象，一方面是由于捡便宜的心理，另一方面是人性的本质导致。通过量化交易编写顺势交易策略可以克服上述人性弱点。
### 1.3.3 避免重仓交易
重仓交易的危害绝不止是让交易者产生重大亏损这么简单。在交易中经常会出现买入后初始阶段并没有盈利的情况，在仓位小的情况下加上合理的止损策略，完全可以等待走势步入正轨。但是过大的仓位会导致资金的大波动，交易者因为无法承受资金的波动选择过早离场，因为间接步入短线的陷阱。另外，还有很多交易者选择向下加仓的结果，即交易者不断加仓已经亏损了很多的头寸，企图通过降低成本来实现交易反转，间接步入逆势的陷阱。
通过量化交易的资金管理模块可以控制交易风险，计算最佳买入开仓量，通过卖出止损因子，可以保证资金的波动在合理范围内。
### 1.3.4 避免对胜率的盲目追求
交易者普遍喜欢追求胜率，希望自己的交易都能以营利收尾，直接表现为：盈利的单子拿不住，稍微有一点利润就忙于兑现；亏损了的交易卖不出，持有到可以再次盈利为止。交易中追求的应该是<b>让利润尽情地奔跑，让亏损尽快地止损</b>，但其实往往变成了<b>让亏损尽情地亏损，让利润尽快的止赢</b>。
通过量化交易，可以制定量化的交易策略和风险控制，通过量化系统的度量模块，可以使交易者拥有正确的交易目标。
### 1.3.5 确保交易策略的执行
在交易中人们往往不能坚持自己的策略决策，因为临时的一点点小变动，将自己的交易搁置甚至反转。通过量化交易，可以确保交易策略的执行，不会因为当天开盘的前的小波动而打乱了原有的交易策略。
### 1.3.6 独立交易及对结果负责的信念
大部分交易者在购入之前喜欢询问他人，如果盈利了，他会认为是自己的能力和本事，如果亏损了，他会把责任推卸到其他人身上。通过量化交易，交易者更容易对结果负责以及培养独立交易的信念。
### 1.3.7 从历史验证交易策略是否可行
量化交易，将需要验证的交易策略进行历史回测，通过对回测结果进行度量即可大致清楚该策略是否有效以及有效范围。
### 1.3.8 寻找交易策略的最优参数
交易中经常见到，自己的交易成交后价格开始下跌，导致止损离场，但随后价格开始不断上涨，导致过早离场。同理止盈点和仓位基准参数都会导致类似的问题。我们希望寻找最优的参数，通过量化交易可以实现最优选择的问题。
### 1.3.9 减少无意义的工作及干扰
通过量化交易，交易者可以腾出很多时间去做其他事情，而不需要无意义的时刻紧盯着行情报价系统，关注价格波动。
