## 4.1 基本操作方法

### 4.1.1 DataFrame构建及方法

```python
#200支股票
stock_cnt=200
#504个交易日
view_days=504
#生成服从正太分布，均值为0，标准差为1的序列
stock_day_change=np.random.standard_normal((stock_cnt,view_days))
stock_day_change
```

```python
pd.DataFrame(stock_day_change)
```

```python
pd.DataFrame(stock_day_change).head()
```

```python
pd.DataFrame(stock_day_change)[5:10]
```

### 4.1.2 索引行列序列

```python
index_list=['股票'+str(i+1) for i in range(stock_day_change.shape[0])]
index_list[::50]
```

```python
pd.DataFrame(stock_day_change,index=index_list)
```

```python
# 构建一组时间
date_list=pd.date_range('2019-1-1',
                        periods=stock_day_change.shape[1],freq='1d')
date_list[55:65]
```

```python
df = pd.DataFrame(stock_day_change,
                  index=index_list,columns=date_list)
df.head()
```

### 4.1.3 金融时间序列

```python
# 矩阵转置
df = df.T
df.head()
```

```python
#采样，每隔21天采样一次
df_21=df.resample('21D',how='mean')
df_21.head()
```

### 4.1.4 Series构建及方法

```python
type(df['股票1'])
#Series就是DataFrame中的一列，可以理解为只有一列的DataFrame
```

```python
plt.subplots(figsize=(14,7))
plt.plot(df['股票1'].cumsum())
```

```python
df_stock_1=df['股票1'].cumsum()
type(df_stock_1)
plt.subplots(figsize=(14,7))
df_stock_1.plot()
```

### 4.1.5 重采样数据

```python
df_stock_1.resample('5D',how='ohlc').plot()
```
